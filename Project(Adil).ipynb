```python
X=df.drop("student_marks",axis="columns")
y=df.drop("study_hours",axis="columns")
print("shape of X = ",X.shape)
print("shape of y = ",y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)

print("shape of X_trian = ",X_train.shape)
print("shape of X_test = ",X_test.shape)
print("shape of y_trian = ",y_train.shape)
print("shape of y_test = ",y_test.shape)
```

# Select a model and train it

```python
# y=m * x + c
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr
```

```python
lr.predict([[11]]).round(2)



#This code is typically used in the context of a machine learning model, specifically a linear regression model (lr) that has been trained using a library like scikit-learn. Here’s what the components mean:

#Components:
#lr: This is the trained linear regression model, often created using LinearRegression() from scikit-learn.

#.predict(): This method is used to make predictions on new input data. In this case, it is being called with [[11]] as the input.

#[[11]]: This is the input data for which the prediction is being made. It's a 2D array (a requirement for scikit-learn models), and 11 represents a feature value for which the prediction is desired.

#.round(2): After making the prediction, the output (which is typically a float) is rounded to 2 decimal places for better readability.
```

```python
lr.predict([[5]])[0][0].round(2)





#lr: This refers to the trained linear regression model, typically from scikit-learn.lr.predic

#.predict([[5]]):

#This is the prediction step. The predict method is called with [[5]] as input.
#[[5]] is a 2D array where 5 represents the value of a feature for which you want to make a prediction. For instance, this could be a feature like "number of hours studied."
#[0][0]:

#The predict() method returns an array of predictions, often a 2D array (e.g., [[value]] for a single prediction).
#The first [0] accesses the first (and likely only) row of predictions.
#The second [0] accesses the single value within that row, effectively extracting the raw prediction as a scalar (a single number).
#.round(2):

#This rounds the scalar prediction value to 2 decimal places for better readability or presentation.
```
